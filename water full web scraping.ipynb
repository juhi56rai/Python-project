```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Fetch the webpage
url = "https://en.wikipedia.org/wiki/List_of_waterfalls_in_India_by_height"
response = requests.get(url)
response.raise_for_status()  # Ensure we got a valid response
print(response)
```

```python
# Parse the HTML
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
# Find the table
table = soup.find('table', class_='wikitable')
table
```

```python
# Extract headers
headers = [th.get_text(strip=True) for th in table.find_all('th')]
headers
```

```python
# Extract rows
rows = []
for tr in table.find_all('tr')[1:]:
    cells = tr.find_all('td')
    row = [cell.get_text(strip=True) for cell in cells]
    rows.append(row)

print(rows)
```

```python
# Create DataFrame
waterfalls_df = pd.DataFrame(rows, columns=headers)

waterfalls_df
```

```python
# Clean and save the data
# Example: Remove empty rows
waterfalls_df.dropna(how='all', inplace=True)

# Save to CSV
waterfalls_df.to_csv('waterfalls_in_india.csv', index=False)
```
